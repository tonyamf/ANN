# Distance signiture Features

## Computing Distance

```python
df["distance"] = ""
for j in range(0, len(df)):
    if len(df["landmarks"][j]) == 0:
        continue
    else:
        xs = []
        ys = []
        df["distance"][j] = []
        for landmark in df["landmarks"][j]:
            for k, v in landmark.items():
                if k == 'chin':
                    continue
                elif k == 'top_lip':
                    xs += v[0][0],  v[3][0], v[6][0]
                    ys += v[0][1],  v[3][1], v[6][1]
                elif k == 'bottom_lip':
                    xs.append(v[3][0])
                    ys.append(v[3][1])
                elif k == 'nose_tip':
                    xs += v[0][0],  v[4][0]
                    ys += v[0][1],  v[4][1]
                elif k == 'nose_bridge':
                    xs.append(v[3][0])
                    ys.append(v[3][1])
                elif k == 'right_eyebrow':
                    xs += v[0][0], v[2][0], v[4][0]
                    ys += v[0][1], v[2][1], v[4][1]
                elif k == 'left_eyebrow':
                    xs += v[0][0], v[2][0], v[4][0]
                    ys += v[0][1], v[2][1], v[4][1]
                else:
                    xs += [x[0] for x in v]
                    ys += [x[1] for x in v]
        for i in range(0, len(ys)-1):
            for s in range(i+1, len(ys)):
                df["distance"][j].append(math.sqrt( pow(xs[i] - xs[s], 2) + pow(ys[i]- ys[s], 2)))
```

```python

df_ =pd.DataFrame(df[df['distance'] != ""].copy())
df_.reset_index(inplace = True, drop = True) 
sums = []
for d in range(0, len(df_["distance"][0])):
    sum = 0
    for i in range(0, len(df_)):
        sum += df_["distance"][i][d]
    sums.append(sum)
```

```python
df_["NormalizedDistance"] = ""
for j in range(0, len(df_)):
    df_["NormalizedDistance"][j] = []
    for d in range(0, len(df_["distance"][j])):
        if sums[d] == 0:
            df_["NormalizedDistance"][j].append(0.0)
        else:
            df_["NormalizedDistance"][j].append(df_["distance"][j][d]/sums[d])
```

## Stability Indes of distance signiture

```python
df_["StabilityIndex_D"] = ""
for j in range(0, len(df_)):
    dif = 0
    for d in range(8, len(df_["NormalizedDistance"][j])):
        dif += abs( df_["NormalizedDistance"][j][d-8] + df_["NormalizedDistance"][j][d] + 8 * (- df_["NormalizedDistance"][j][d-1] 
        - df_["NormalizedDistance"][j][d-2] - df_["NormalizedDistance"][j][d-3] - df_["NormalizedDistance"][j][d-4] 
        - df_["NormalizedDistance"][j][d-5] - df_["NormalizedDistance"][j][d-6] - df_["NormalizedDistance"][j][d-7]) ) 
    df_["StabilityIndex_D"][j] = abs(dif)
```

## Distance statictic

```python
df_["Skewness_D"] = ""
df_["Kurtosis_D"] = ""
df_["Moment_D"] = ""
df_["Entropy_D"] = ""
for j in range(0, len(df_)):
    df_["Moment_D"][j] = []
    stats = scipy.stats.describe(df_["NormalizedDistance"][j])
    df_["Skewness_D"][j] = stats.skewness
    df_["Kurtosis_D"] = stats.kurtosis
    df_["Moment_D"][j].append(scipy.stats.moment(df_["NormalizedDistance"][j], moment=1))
    df_["Moment_D"][j].append(scipy.stats.moment(df_["NormalizedDistance"][j], moment=2))
    df_["Moment_D"][j].append(scipy.stats.moment(df_["NormalizedDistance"][j], moment=3))
    df_["Moment_D"][j].append(scipy.stats.moment(df_["NormalizedDistance"][j], moment=4))
    df_["Entropy_D"] = scipy.stats.entropy(df_["NormalizedDistance"][j])
```

# Shape signiture Features

## Compute shape signiture

```python
df_["shape"] = ""
for j in range(0, len(df_)):
    xs = []
    ys = []
    df_["shape"][j] = []
    for landmark in df_["landmarks"][j]:
        for k, v in landmark.items():
            if k == 'chin':
                continue
            elif k == 'top_lip':
                xs += v[0][0],  v[3][0], v[6][0]
                ys += v[0][1],  v[3][1], v[6][1]
            elif k == 'bottom_lip':
                xs.append(v[3][0])
                ys.append(v[3][1])
            elif k == 'nose_tip':
                xs += v[0][0],  v[4][0]
                ys += v[0][1],  v[4][1]
            elif k == 'nose_bridge':
                xs.append(v[3][0])
                ys.append(v[3][1])
            elif k == 'right_eyebrow':
                xs += v[0][0], v[2][0], v[4][0]
                ys += v[0][1], v[2][1], v[4][1]
            elif k == 'left_eyebrow':
                xs += v[0][0], v[2][0], v[4][0]
                ys += v[0][1], v[2][1], v[4][1]
            else:
                xs += [x[0] for x in v]
                ys += [x[1] for x in v]
    for i in range(0, len(ys)-2):
        for s in range(i+1, len(ys)-1):
            for z in range(s+1, len(ys)):
                a = (xs[i] * (ys[s] - ys[z]) +
                     xs[s] * (ys[z] - ys[i]) + 
                     xs[z] * (ys[i] - ys[s])) 
                if a != 0: 
                    ab = math.sqrt( pow(xs[i] - xs[s], 2) + pow(ys[i]- ys[s], 2))
                    bc = math.sqrt( pow(xs[s] - xs[z], 2) + pow(ys[s]- ys[z], 2))
                    ca = math.sqrt( pow(xs[z] - xs[i], 2) + pow(ys[z]- ys[i], 2))
                    h = (ab + bc + ca)/2
                    result = math.sqrt( h * (h-ab) * (h-bc) *(h-ca)) / pow(h*2, 2)
                    df_["shape"][j].append(result)
```
